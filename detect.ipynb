```python
from ultralytics import YOLO
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python

# carrega o modelo
model = YOLO(r"C:/Users/Gesser/Documents/BRUNO/yolo custom/train9/weights/best.pt")  # modelo yolo v8 pretreinado

# Run batched inference on a list of images
image_path = r"C:/Users/Gesser/Documents/BRUNO/dji images/crop2.png"
source = cv2.imread( image_path )

source = cv2.resize(source,  (0, 0), cv2.INTER_LANCZOS4, fx = 0.5, fy = 0.5) #INTER_LANCZOS4 guarda melhor qualidade dos pixels

model.predict(source, imgsz = len(source), save=True)  # retorna a lsita de resultados obtidos

results = model( image_path )

# processo para retornar lista
for result in results:
    '''boxes = result.boxes
    masks = result.masks
    keypoints = result.keypoints
    probs = result.probs
    obb = result.obb'''
    result.show()  # mostra a tela
    result.save(filename="result.jpg")  # salva o arquivo
```
